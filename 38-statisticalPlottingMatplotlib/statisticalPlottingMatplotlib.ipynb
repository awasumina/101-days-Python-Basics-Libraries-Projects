```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already defined mens_rowing and mens_gymnastics DataFrames
mens_rowing = pd.DataFrame({
    "Name": ["Giovanni Abagnale", "Jrmie Azou", "Thomas Gabriel Jrmie Baroukh"],
    "Height": [190, 185, 175]  # Example heights for demonstration
})

mens_gymnastics = pd.DataFrame({
    "Name": ["Denis Mikhaylovich Ablyazin", "David Sagitovich Belyavsky", "Fabian Hambchen"],
    "Height": [170, 175, 165]  # Example heights for demonstration
})

# Calculate mean and standard deviation for "Height" column in mens_rowing
mean_height_rowing = mens_rowing['Height'].mean()
std_height_rowing = mens_rowing['Height'].std()

# Calculate mean and standard deviation for "Height" column in mens_gymnastics
mean_height_gymnastics = mens_gymnastics['Height'].mean()
std_height_gymnastics = mens_gymnastics['Height'].std()

# Create a figure and axis
fig, ax = plt.subplots()

# Add a bar for the rowing "Height" column mean/std
ax.bar("Rowing", mean_height_rowing, yerr=std_height_rowing, label="Rowing")

# Add a bar for the gymnastics "Height" column mean/std
ax.bar("Gymnastics", mean_height_gymnastics, yerr=std_height_gymnastics, label="Gymnastics")

# Label the y-axis
ax.set_ylabel("Height (cm)")

# Add a title
ax.set_title("Mean and Standard Deviation of Height")

# Add a legend
ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define Seattle weather data
seattle_weather = pd.DataFrame({
    "MONTH": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "MLY-TAVG-NORMAL": [41.8, 44.2, 47.7, 51.9, 57.2, 61.5, 65.5, 65.6, 60.5, 52.5, 45.3, 41.1],
    "MLY-TAVG-STDDEV": [2.3, 2.6, 1.8, 1.5, 1.9, 1.2, 1.2, 1.3, 1.5, 1.8, 3.0, 2.3]
})

# Define Austin weather data
austin_weather = pd.DataFrame({
    "MONTH": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "MLY-TAVG-NORMAL": [49.6, 53.2, 61.7, 69.9, 76.9, 83.1, 85.6, 86.0, 80.9, 70.9, 60.0, 50.8],
    "MLY-TAVG-STDDEV": [2.9, 3.2, 3.6, 4.1, 2.5, 1.9, 1.3, 1.4, 2.5, 3.7, 2.9, 4.0]
})

# Create a figure and axis
fig, ax = plt.subplots()     

# Add Seattle temperature data in each month with error bars
ax.errorbar(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'], yerr=seattle_weather["MLY-TAVG-STDDEV"], label='Seattle')

# Add Austin temperature data in each month with error bars
ax.errorbar(austin_weather['MONTH'], austin_weather['MLY-TAVG-NORMAL'], yerr=austin_weather["MLY-TAVG-STDDEV"], label='Austin')

# Set the y-axis label
ax.set_ylabel("Temperature (Fahrenheit)")

# Add a legend
ax.legend()

plt.show()
```
