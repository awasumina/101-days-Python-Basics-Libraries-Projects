```python
# Histograms allow us to see the distributions of the data in different groups in our data. In this exercise, you will select groups from the Summer 2016 Olympic Games medalist dataset to compare the height of medalist athletes in two different sports



import pandas as pd
import matplotlib.pyplot as plt

# Define the data for mens_rowing
mens_rowing_data = {
    'Name': ['Giovanni Abagnale', 'Jrmie Azou', 'Thomas Gabriel Jrmie Baroukh', 'Jacob Jepsen Barse', 'Alexander Belonogoff', 'Giuseppe Vicino', 'Phillipp Wende', 'Peter Noel Jan Wiersum', 'Tone Wieten', 'Kasper Winther Jrgensen'],
    'Weight': [90, 70, 80, 85, 75, 78, 85, 95, 88, 83]
}
mens_rowing = pd.DataFrame(mens_rowing_data)

# Define the data for mens_gymnastics
mens_gymnastics_data = {
    'Name': ['Denis Mikhaylovich Ablyazin', 'David Sagitovich Belyavsky', 'Deng Shudi', 'Fabian Hambchen', 'Diego Matias Hyplito', 'Ryohei Kato', 'Nikolay Yulyevich Kuksenkov', 'Danell Johan Leyva Gonzalez', 'Lin Chaopan', 'Liu Yang'],
    'Weight': [65, 68, 70, 73, 72, 66, 68, 69, 67, 64]
}
mens_gymnastics = pd.DataFrame(mens_gymnastics_data)

fig, ax = plt.subplots()

# Plot a histogram of "Weight" for mens_rowing
ax.hist(mens_rowing['Weight'],label="Rowing",histtype="step",bins=5)

# Compare to histogram of "Weight" for mens_gymnastics
ax.hist(mens_gymnastics['Weight'],label="Gymnastics",histtype="step",bins=5)

ax.set_xlabel("Weight (kg)")
ax.set_ylabel("# of observations")

# Add the legend and show the Figure
ax.legend()
plt.show()
```
