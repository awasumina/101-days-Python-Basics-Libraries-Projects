```python
import pandas as pd

data = {
    'date': ['2015-12-27', '2015-12-20', '2015-12-13', '2015-12-06', '2015-11-29', '2018-02-04', '2018-01-28', '2018-01-21', '2018-01-14', '2018-01-07'],
    'type': ['conventional', 'conventional', 'conventional', 'conventional', 'conventional', 'organic', 'organic', 'organic', 'organic', 'organic'],
    'year': [2015, 2015, 2015, 2015, 2015, 2018, 2018, 2018, 2018, 2018],
    'avg_price': [0.95, 0.98, 0.93, 0.89, 0.99, 1.53, 1.61, 1.63, 1.59, 1.51],
    'size': ['small', 'small', 'small', 'small', 'small', 'extra_large', 'extra_large', 'extra_large', 'extra_large', 'extra_large'],
    'nb_sold': [9627000, 8710000, 9855000, 9405000, 8095000, 1704, 1271, 1490, 1580, 1289]
}

avocados = pd.DataFrame(data)

# Import matplotlib.pyplot with alias plt
import matplotlib.pyplot as plt


# Modify bins to 20
avocados[avocados["type"] == "conventional"]["avg_price"].hist(alpha=0.5,bins=20)

# Modify bins to 20
avocados[avocados["type"] == "organic"]["avg_price"].hist(alpha=0.5,bins=20)

# Add a legend
plt.legend(["conventional", "organic"])

# Show the plot
plt.show()

```
