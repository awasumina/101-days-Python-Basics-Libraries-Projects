```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = {
    'school': ['GP', 'GP', 'GP', 'GP', 'GP', 'MS', 'MS', 'MS', 'MS', 'MS'],
    'sex': ['F', 'F', 'F', 'F', 'F', 'M', 'M', 'M', 'M', 'M'],
    'age': [18, 17, 15, 15, 16, 20, 17, 21, 18, 19],
    'famsize': ['GT3', 'GT3', 'LE3', 'GT3', 'GT3', 'LE3', 'LE3', 'GT3', 'LE3', 'LE3'],
    'Pstatus': ['A', 'T', 'T', 'T', 'T', 'A', 'T', 'T', 'T', 'T'],
    'G1': [5, 5, 7, 15, 6, 9, 14, 10, 11, 8],
    'G2': [6, 5, 8, 14, 10, 9, 16, 8, 12, 9],
    'G3': [6, 6, 10, 15, 10, 9, 16, 7, 10, 9],
    'location': ['Rural', 'Urban', 'Rural', 'Urban', 'Urban', 'Urban', 'Urban', 'Rural', 'Rural', 'Urban'],
    'study_time': ['2 to 5 hours', '2 to 5 hours', '2 to 5 hours', '5 to 10 hours', '2 to 5 hours',
                   '2 to 5 hours', '<2 hours', '<2 hours', '<2 hours', '<2 hours']
}


student_data = pd.DataFrame(data)



# Create a dictionary mapping subgroup values to colors
palette_colors = {'Rural': "green", 'Urban': "blue"}

# Create a count plot of school with location subgroups
sns.countplot(x="school",data=student_data,hue="location",palette=palette_colors)



# Display plot
plt.show()
```
