```python
import matplotlib.pyplot as plt
import pandas as pd

# Define the climate_change DataFrame
data = {
    'date': ['1958-03-06', '1958-04-06', '1958-05-06', '1958-06-06', '1958-07-06'],
    'co2': [315.71, 317.45, 317.5, float('nan'), 315.86],
    'relative_temp': [0.1, 0.01, 0.08, -0.05, 0.06]
}
climate_change = pd.DataFrame(data)
climate_change['date'] = pd.to_datetime(climate_change['date'])
climate_change.set_index('date', inplace=True)


fig, ax = plt.subplots()

# Add data: "co2" on x-axis, "relative_temp" on y-axis
ax.scatter(climate_change['co2'],climate_change['relative_temp'])

# Set the x-axis label to "CO2 (ppm)"
ax.set_xlabel('CO2 (ppm)')

# Set the y-axis label to "Relative temperature (C)"
ax.set_ylabel('Relative temperature (C)')


plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Define the climate_change DataFrame
data = {
    'date': ['1958-03-06', '1958-04-06', '1958-05-06', '1958-06-06', '1958-07-06'],
    'co2': [315.71, 317.45, 317.5, float('nan'), 315.86],
    'relative_temp': [0.1, 0.01, 0.08, -0.05, 0.06]
}
climate_change = pd.DataFrame(data)
climate_change['date'] = pd.to_datetime(climate_change['date'])
climate_change.set_index('date', inplace=True)


fig, ax = plt.subplots()

# Add data: "co2", "relative_temp" as x-y, index as color
ax.scatter(climate_change['co2'],climate_change['relative_temp'],c=climate_change.index)


# Set the x-axis label to "CO2 (ppm)"
ax.set_xlabel('CO2 (ppm)')


# Set the y-axis label to "Relative temperature (C)"
ax.set_ylabel('Relative temperature (C)')


plt.show()
```
