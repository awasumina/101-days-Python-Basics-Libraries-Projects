```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd

# Define the data for the dataframe
data = {
    'fips2010': [100199999, 100399999, 100599999, 100799999, 100999999,
                 7215199999, 7215399999, 7801099999, 7802099999, 7803099999],
    'fmr_0': [588, 729, 491, 692, 692, 415, 323, 735, 977, 703],
    'fmr_1': [691, 762, 494, 768, 768, 439, 337, 750, 1158, 848],
    'fmr_2': [829, 879, 657, 882, 882, 530, 389, 916, 1439, 1061],
    'fmr_3': [1055, 1280, 824, 1176, 1176, 706, 501, 1148, 1804, 1330],
    'fmr_4': [1234, 1456, 935, 1289, 1289, 850, 605, 1300, 2021, 1498],
    'state_alpha': ['AL', 'AL', 'AL', 'AL', 'AL', 'PR', 'PR', 'VI', 'VI', 'VI'],
    'fmr_type': [40, 40, 40, 40, 40, 40, 40, 40, 40, 40],
    'metro': [1, 1, 0, 1, 1, 1, 1, 0, 0, 0],
    'fmr_pct_chg': [0.009, -0.100, -0.021, 0.018, 0.018, -0.042, -0.100, 0.007, 0.006, 0.007],
    'fmr_dollar_chg': [7, -98, -14, 16, 16, -23, -43, 6, 9, 7]
}

# Create the dataframe
df = pd.DataFrame(data)

# Loop through differences between bright and colorblind palettes
for p in ['bright', 'colorblind']:
    sns.set_palette(p)
    sns.displot(df['fmr_3'])
    plt.show()

    # Clear the plots
    plt.clf()
```
