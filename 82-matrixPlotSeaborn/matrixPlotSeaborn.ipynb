```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the data for the dataframe
data = {
    'YEAR': [
        1999, 1999, 1999, 1999, 1999,
        # Add more years as needed
    ],
    'GoogleKnowlege_Occupation': [
        'actor', 'Comedian', 'television actress', 'film actress', 'actor',
        # Add more occupations as needed
    ],
    'Show': [
        '1/11/99', '1/12/99', '1/13/99', '1/14/99', '1/18/99',
        # Add more show dates as needed
    ],
    'Group': [
        'Acting', 'Comedy', 'Acting', 'Acting', 'Acting',
        # Add more groups as needed
    ],
    'Raw_Guest_List': [
        'Michael J. Fox', 'Sandra Bernhard', 'Tracey Ullman', 'Gillian Anderson', 'David Alan Grier',
        # Add more guest names as needed
    ]
}

# Create the dataframe
df = pd.DataFrame(data)

# Create the crosstab DataFrame
pd_crosstab = pd.crosstab(df["Group"], df["YEAR"])

# Plot a heatmap of the table with no color bar and using the BuGn palette
sns.heatmap(pd_crosstab, cbar=False, cmap="BuGn", linewidths=.3)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)

#Show the plot
plt.show()
plt.clf()
```
