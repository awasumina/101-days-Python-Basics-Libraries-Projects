```python
import pandas as pd

data = {
    'date': [
        '2000-01-01', '2000-02-01', '2000-03-01', '2000-04-01', '2000-05-01',
        '2013-05-01', '2013-06-01', '2013-07-01', '2013-08-01', '2013-09-01'
    ],
    'city': [
        'Abidjan', 'Abidjan', 'Abidjan', 'Abidjan', 'Abidjan',
        'Xian', 'Xian', 'Xian', 'Xian', 'Xian'
    ],
    'country': [
        "Côte D'Ivoire", "Côte D'Ivoire", "Côte D'Ivoire", "Côte D'Ivoire", "Côte D'Ivoire",
        'China', 'China', 'China', 'China', 'China'
    ],
    'avg_temp_c': [
        27.293, 27.685, 29.061, 28.162, 27.547,
        18.979, 23.522, 25.251, 24.528, None
    ]
}

temperatures = pd.DataFrame(data)



# Index temperatures by country & city
temperatures_ind = temperatures.set_index(['country','city'])

# List of tuples: Brazil, Rio De Janeiro & Pakistan, Lahore

rows_to_keep = [("Brazil", "Rio De Janeiro"), ("Pakistan", "Lahore")]

# Subset for rows to keep
print(temperatures_ind.loc[rows_to_keep])
```
